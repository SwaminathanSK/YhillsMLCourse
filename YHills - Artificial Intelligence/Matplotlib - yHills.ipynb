```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
### Draw a line in a diagram from pos (1, 5) to (2, 10)
xpoints = np.array([1, 2])
ypoints = np.array([5, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 2])
ypoints = np.array([5, 10])

plt.plot(xpoints, ypoints, '+')
plt.show()
```

```python
ypoints = np.array([3, 8, 10])
plt.plot(ypoints, '*'':y')
plt.show()
```

```python
# : makes it dotted lines, y makes it yellow
```

### marker, linestyle, color, ms, mfc, mec are arguments we can use in the plot() function. where ms = marker size, mfc = marker face color, mec = marker edge color

```python
ypoints = np.array([3, 8, 10])
plt.plot(ypoints, '*'':y')
plt.title('bruh')
plt.xlabel('xnega')
plt.ylabel('yoosss')
plt.show()
```

```python
xpoints = np.array([1, 2, 3, 6])
ypoints = np.array([3, 5, 6, 8])
plt.scatter(xpoints, ypoints)
plt.show()
```

```python
plt.pie(ypoints)
```

```python
plt.bar(xpoints, ypoints)
```

```python
plt.hist(xpoints, ypoints)
plt.show()
```
