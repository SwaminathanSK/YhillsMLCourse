```python
### Data Visualization
categorical_columns = ['age_bracket', 'qualification', 'race', 'sex', 'income_level', 'marital_status', 'housing_status', 'employment']

fig,axes = plt.subplots(4, 2,figsize=(16,20))
for idx,cat_col in enumerate(categorical_columns):
    row,col = idx//2,idx%2
    sns.countplot(x=cat_col,data=df,hue='h1n1_vaccine',ax=axes[row,col])


plt.subplots_adjust(hspace=1)
```

```python
sns.countplot(x = 'census_msa', data = df, hue = 'h1n1_vaccine')
```

### From the above plots we can observe the following:

#### 1. Most of the population in different demographics have not taken the H1N1 vaccine.
#### 2. The data contains mostly of white race and less than 30 percent of the white population have had the vaccine.
#### 3. There is a greater count of female population who had taken the vaccine. But this could also be due to a greater count of the total female population in the data
#### 4. Majority of the data points are taken from Metropolitan Statistical Area, Non-principle city (Not the largest city in the MSA).
#### 5. People with own houses are more than rented houses in the dataset. And about 1/5th of the people with own house have had the vaccine.
#### 6. The dataset has more people who are married. But it appears that a slightly greater percentage of the non-married people have had the vaccine.
#### 7. Dataset contains datapoints of employed citizens in majority, but the percentage of those who had the vaccine is less.
